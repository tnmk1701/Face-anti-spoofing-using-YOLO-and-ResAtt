```python
import cv2
import os
import glob
import numpy as np
from ultralytics import YOLO
from ultralytics.utils.metrics import bbox_iou

def calculate_iou(box1, box2):
    # Chuyá»ƒn Ä‘á»•i format bbox náº¿u cáº§n
    x1, y1, x2, y2 = box1
    x3, y3, x4, y4 = box2

    # TÃ­nh toáº¡ Ä‘á»™ giao nhau
    x_left = max(x1, x3)
    y_top = max(y1, y3)
    x_right = min(x2, x4)
    y_bottom = min(y2, y4)

    # TÃ­nh diá»‡n tÃ­ch giao nhau
    intersection_area = max(0, x_right - x_left) * max(0, y_bottom - y_top)

    # TÃ­nh diá»‡n tÃ­ch hai bbox
    box1_area = (x2 - x1) * (y2 - y1)
    box2_area = (x4 - x3) * (y4 - y3)

    # TÃ­nh IoU
    union_area = box1_area + box2_area - intersection_area
    iou = intersection_area / union_area if union_area > 0 else 0

    return iou

def detect_faces_haar(image_path):
    # Load Haar Cascade classifier
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    # Äá»c áº£nh
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # Convert to YOLO bbox format (x1, y1, x2, y2)
    haar_boxes = []
    for (x, y, w, h) in faces:
        haar_boxes.append([x, y, x+w, y+h])

    return haar_boxes

def compare_detection_methods(yolo_model, test_dir):
    # Láº¥y danh sÃ¡ch áº£nh
    image_paths = glob.glob(os.path.join(test_dir, '*.[jpngbmp]*'))

    # LÆ°u káº¿t quáº£
    results = {
        'yolo_iou': [],
        'haar_iou': [],
        'total_yolo_faces': 0,
        'total_haar_faces': 0
    }

    for img_path in image_paths:
        # YOLO detection
        yolo_results = yolo_model(img_path)
        yolo_boxes = yolo_results[0].boxes.xyxy.cpu().numpy()
        results['total_yolo_faces'] += len(yolo_boxes)

        # Haar Cascade detection
        haar_boxes = detect_faces_haar(img_path)
        results['total_haar_faces'] += len(haar_boxes)

        # TÃ­nh IoU
        if len(yolo_boxes) > 0 and len(haar_boxes) > 0:
            # TÃ­nh IoU giá»¯a cÃ¡c bounding box
            iou_matrix = np.zeros((len(yolo_boxes), len(haar_boxes)))

            for i, yolo_box in enumerate(yolo_boxes):
                for j, haar_box in enumerate(haar_boxes):
                    iou_matrix[i, j] = calculate_iou(yolo_box, haar_box)

            # Láº¥y IoU max
            max_iou_yolo = np.max(iou_matrix, axis=1).mean()
            max_iou_haar = np.max(iou_matrix, axis=0).mean()

            results['yolo_iou'].append(max_iou_yolo)
            results['haar_iou'].append(max_iou_haar)

    # TÃ­nh toÃ¡n káº¿t quáº£
    results['mean_yolo_iou'] = np.mean(results['yolo_iou']) if results['yolo_iou'] else 0
    results['mean_haar_iou'] = np.mean(results['haar_iou']) if results['haar_iou'] else 0

    return results

# Load YOLO model
# model = YOLO('path/to/your/best.pt')

# ThÆ° má»¥c test
test_dir = "/content/drive/MyDrive/FaceAntiSpoofing/data/test/images"

# So sÃ¡nh
comparison_results = compare_detection_methods(model, test_dir)

# In káº¿t quáº£
print("Comparison Results:")
print(f"Total YOLO Detected Faces: {comparison_results['total_yolo_faces']}")
print(f"Total Haar Cascade Detected Faces: {comparison_results['total_haar_faces']}")
print(f"Mean YOLO IoU: {comparison_results['mean_yolo_iou']:.4f}")
print(f"Mean Haar Cascade IoU: {comparison_results['mean_haar_iou']:.4f}")
```

```python
import os
import numpy as np
import cv2
from ultralytics import YOLO

def load_yolo_annotations(txt_path):
    """Äá»c annotation YOLO format"""
    with open(txt_path, 'r') as f:
        lines = f.readlines()

    # Chuyá»ƒn Ä‘á»•i YOLO format sang pixel coordinates
    boxes = []
    for line in lines:
        parts = line.strip().split()
        class_id = int(parts[0])
        x_center, y_center, width, height = map(float, parts[1:])

        # Äá»c kÃ­ch thÆ°á»›c áº£nh
        img_path = txt_path.replace('.txt', '.jpg')
        img = cv2.imread(img_path)
        img_height, img_width = img.shape[:2]

        # Chuyá»ƒn vá» pixel coordinates
        x1 = int((x_center - width/2) * img_width)
        y1 = int((y_center - height/2) * img_height)
        x2 = int((x_center + width/2) * img_width)
        y2 = int((y_center + height/2) * img_height)

        boxes.append([x1, y1, x2, y2])

    return boxes

def calculate_iou_with_ground_truth(test_dir, model):
    # Láº¥y danh sÃ¡ch áº£nh
    image_paths = glob.glob(os.path.join(test_dir, '*.[jpngbmp]*'))

    iou_scores = []
    for img_path in image_paths:
        # Äá»c ground truth
        txt_path = img_path.replace('.jpg', '.txt')
        if not os.path.exists(txt_path):
            continue

        # Ground truth boxes
        gt_boxes = load_yolo_annotations(txt_path)

        # Predict boxes tá»« model
        results = model(img_path)
        pred_boxes = results[0].boxes.xyxy.cpu().numpy()

        # TÃ­nh IoU
        img_iou_scores = []
        for pred_box in pred_boxes:
            # TÃ­nh IoU vá»›i má»—i ground truth box
            max_iou = max(calculate_iou(pred_box, gt_box) for gt_box in gt_boxes)
            img_iou_scores.append(max_iou)

        # Trung bÃ¬nh IoU cá»§a áº£nh nÃ y
        if img_iou_scores:
            iou_scores.append(np.mean(img_iou_scores))

    # Trung bÃ¬nh IoU toÃ n bá»™
    mean_iou = np.mean(iou_scores) if iou_scores else 0

    return mean_iou, iou_scores

# Load model
model = YOLO('/content/drive/MyDrive/FaceAntiSpoofing/model_train_yolo/train/weights/best.pt')

# ThÆ° má»¥c test
test_dir = "/content/drive/MyDrive/FaceAntiSpoofing/data/test/images"

# TÃ­nh IoU
mean_iou, individual_ious = calculate_iou_with_ground_truth(test_dir, model)
print(f"Mean IoU: {mean_iou}")
print("Individual IoUs:", individual_ious)
```

```python
import os
import cv2
import numpy as np
from ultralytics import YOLO

def draw_boxes_on_image(img_path, gt_boxes, pred_boxes, output_dir):
    # Äá»c áº£nh
    img = cv2.imread(img_path)

    # Váº½ ground truth boxes (mÃ u xanh)
    for box in gt_boxes:
        x1, y1, x2, y2 = map(int, box)
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(img, 'GT', (x1, y1-10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Váº½ predicted boxes (mÃ u Ä‘á»)
    for box in pred_boxes:
        x1, y1, x2, y2 = map(int, box)
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 0, 255), 2)
        cv2.putText(img, 'Pred', (x1, y1-10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)

    # Táº¡o thÆ° má»¥c output náº¿u chÆ°a tá»“n táº¡i
    os.makedirs(output_dir, exist_ok=True)

    # LÆ°u áº£nh
    output_filename = os.path.basename(img_path)
    output_path = os.path.join(output_dir, output_filename)
    cv2.imwrite(output_path, img)

    return output_path

def visualize_detection_results(test_images_dir, test_labels_dir, model, output_dir):
    # Láº¥y danh sÃ¡ch áº£nh
    image_paths = [f for f in os.listdir(test_images_dir) if f.endswith(('.jpg', '.png', '.jpeg'))]

    visualized_images = []

    for img_filename in image_paths:
        # ÄÆ°á»ng dáº«n Ä‘áº§y Ä‘á»§
        img_path = os.path.join(test_images_dir, img_filename)

        # TÃ¬m file annotation tÆ°Æ¡ng á»©ng
        txt_filename = os.path.splitext(img_filename)[0] + '.txt'
        txt_path = os.path.join(test_labels_dir, txt_filename)

        # Kiá»ƒm tra file annotation tá»“n táº¡i
        if not os.path.exists(txt_path):
            print(f"No annotation found for {img_filename}")
            continue

        # Äá»c ground truth boxes
        gt_boxes = []
        with open(txt_path, 'r') as f:
            lines = f.readlines()

        # Äá»c kÃ­ch thÆ°á»›c áº£nh
        img = cv2.imread(img_path)
        img_height, img_width = img.shape[:2]

        # Chuyá»ƒn Ä‘á»•i YOLO format sang pixel coordinates
        for line in lines:
            parts = line.strip().split()
            class_id = int(parts[0])
            x_center, y_center, width, height = map(float, parts[1:])

            x1 = int((x_center - width/2) * img_width)
            y1 = int((y_center - height/2) * img_height)
            x2 = int((x_center + width/2) * img_width)
            y2 = int((y_center + height/2) * img_height)

            gt_boxes.append([x1, y1, x2, y2])

        # Predict boxes tá»« model
        results = model(img_path)
        pred_boxes = results[0].boxes.xyxy.cpu().numpy()

        # Váº½ boxes lÃªn áº£nh
        output_path = draw_boxes_on_image(
            img_path,
            gt_boxes,
            pred_boxes,
            output_dir
        )

        visualized_images.append(output_path)

    return visualized_images

# Load model
model = YOLO('/content/drive/MyDrive/FaceAntiSpoofing/model_train_yolo/train/weights/best.pt')

# ThÆ° má»¥c Ä‘áº§u vÃ o vÃ  Ä‘áº§u ra
test_images_dir = "/content/drive/MyDrive/FaceAntiSpoofing/data/test/images"
test_labels_dir = "/content/drive/MyDrive/FaceAntiSpoofing/data/test/labels"
output_dir = "/content/drive/MyDrive/FaceAntiSpoofing/visualization_results"

# Visualize káº¿t quáº£
visualized_images = visualize_detection_results(
    test_images_dir,
    test_labels_dir,
    model,
    output_dir
)

# In ra Ä‘Æ°á»ng dáº«n cÃ¡c áº£nh Ä‘Ã£ visualize
print("Visualized images:")
for img in visualized_images:
    print(img)
```

```python
import os
import glob
import numpy as np
import cv2
import time  # ThÃªm thÆ° viá»‡n Ä‘o thá»i gian

def calculate_iou(box1, box2):
    x1, y1, x2, y2 = box1
    x3, y3, x4, y4 = box2

    x_left = max(x1, x3)
    y_top = max(y1, y3)
    x_right = min(x2, x4)
    y_bottom = min(y2, y4)

    intersection_area = max(0, x_right - x_left) * max(0, y_bottom - y_top)
    box1_area = (x2 - x1) * (y2 - y1)
    box2_area = (x4 - x3) * (y4 - y3)

    union_area = box1_area + box2_area - intersection_area
    iou = intersection_area / union_area if union_area > 0 else 0

    return iou

def load_yolo_annotations(txt_path, img_path):
    with open(txt_path, 'r') as f:
        lines = f.readlines()

    boxes = []
    for line in lines:
        parts = line.strip().split()
        class_id = int(parts[0])
        x_center, y_center, width, height = map(float, parts[1:])

        img = cv2.imread(img_path)
        img_height, img_width = img.shape[:2]

        x1 = int((x_center - width/2) * img_width)
        y1 = int((y_center - height/2) * img_height)
        x2 = int((x_center + width/2) * img_width)
        y2 = int((y_center + height/2) * img_height)

        boxes.append([x1, y1, x2, y2])

    return boxes

def calculate_iou_with_haarcascade(test_images_dir, test_labels_dir, haarcascade_path):
    # Load Haar Cascade model
    face_cascade = cv2.CascadeClassifier(haarcascade_path)

    # Láº¥y danh sÃ¡ch áº£nh
    image_paths = glob.glob(os.path.join(test_images_dir, '*.[jpngbmp]*'))

    iou_scores = []
    detailed_results = {
        'images_processed': 0,
        'total_gt_boxes': 0,
        'total_pred_boxes': 0,
        'iou_details': []
    }

    start_time = time.time()  # Báº¯t Ä‘áº§u Ä‘o runtime toÃ n bá»™ quÃ¡ trÃ¬nh

    for img_path in image_paths:
        img_start_time = time.time()  # Báº¯t Ä‘áº§u Ä‘o runtime cho tá»«ng áº£nh

        img_filename = os.path.basename(img_path)
        txt_filename = os.path.splitext(img_filename)[0] + '.txt'
        txt_path = os.path.join(test_labels_dir, txt_filename)

        if not os.path.exists(txt_path):
            print(f"No annotation found for {img_filename}")
            continue

        gt_boxes = load_yolo_annotations(txt_path, img_path)
        detailed_results['total_gt_boxes'] += len(gt_boxes)

        # Dá»± Ä‘oÃ¡n báº±ng Haar Cascade
        img = cv2.imread(img_path)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        detected_boxes = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

        pred_boxes = []
        for (x, y, w, h) in detected_boxes:
            pred_boxes.append([x, y, x + w, y + h])
        detailed_results['total_pred_boxes'] += len(pred_boxes)

        img_iou_scores = []
        matched_boxes = []

        for pred_box in pred_boxes:
            iou_with_gts = [calculate_iou(pred_box, gt_box) for gt_box in gt_boxes]
            if iou_with_gts:
                max_iou = max(iou_with_gts)
                img_iou_scores.append(max_iou)
                matched_boxes.append({
                    'pred_box': pred_box,
                    'max_iou': max_iou,
                    'matched_gt_box': gt_boxes[iou_with_gts.index(max_iou)]
                })

        if img_iou_scores:
            img_mean_iou = np.mean(img_iou_scores)
            iou_scores.append(img_mean_iou)

            detailed_results['iou_details'].append({
                'image': img_filename,
                'mean_iou': img_mean_iou,
                'matched_boxes': matched_boxes
            })

        detailed_results['images_processed'] += 1

        img_end_time = time.time()  # Káº¿t thÃºc Ä‘o runtime cho tá»«ng áº£nh
        print(f"Processed {img_filename} in {img_end_time - img_start_time:.2f} seconds")

    end_time = time.time()  # Káº¿t thÃºc Ä‘o runtime toÃ n bá»™ quÃ¡ trÃ¬nh

    mean_iou = np.mean(iou_scores) if iou_scores else 0

    print("\nDetailed Results:")
    print(f"Images Processed: {detailed_results['images_processed']}")
    print(f"Total Ground Truth Boxes: {detailed_results['total_gt_boxes']}")
    print(f"Total Predicted Boxes: {detailed_results['total_pred_boxes']}")
    print(f"Mean IoU: {mean_iou:.4f}")
    print(f"Total Runtime: {end_time - start_time:.2f} seconds")

    return mean_iou, iou_scores, detailed_results

# ÄÆ°á»ng dáº«n Haar Cascade XML
haarcascade_path = cv2.data.haarcascades + "haarcascade_frontalface_default.xml"

# ThÆ° má»¥c test
test_images_dir = "/content/drive/MyDrive/FaceAntiSpoofing/data/test/images"
test_labels_dir = "/content/drive/MyDrive/FaceAntiSpoofing/data/test/labels"

# TÃ­nh IoU
mean_iou, individual_ious, detailed_results = calculate_iou_with_haarcascade(
    test_images_dir,
    test_labels_dir,
    haarcascade_path
)

# In IoU cá»§a tá»«ng áº£nh
print("\nIndividual IoUs:", individual_ious)
```
